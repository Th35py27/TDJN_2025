#### Q5. What will be the output in the code segment below?

```java
public class Main {
    public static void main(String[] args) {
        int[][] scores = {
            {90, 85, 78},
            {75, 88, 92},
            {88, 90, 85},
            {95, 80, 70}
        };

        int totalScore = 0;

        for (int[] studentScores : scores) {
            for (int score : studentScores) {
                totalScore += score;
            }
        }

        System.out.println("Total Score: " + totalScore);
    }
}

```

B. 348
<br>
